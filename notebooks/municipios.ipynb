```python
import json
import pandas as pd
from pathlib import Path
from sina.config.paths import GAS_DATA
```

```python
CSV_PATH = GAS_DATA / Path("municipios.csv")
df_municipios = pd.read_csv(CSV_PATH)
```

```python
df_municipios.rename(columns={
    "_id": "ClaveEntidad",
    "Estado": "EntidadFederativa",
    "CodMunicipio": "ClaveMunicipio",
    "MUNICIPIO": "Municipio",
}, inplace=True)
```

```python
df_municipios.info()
```

```python
from unidecode import unidecode
def limpiar_texto(texto: str) -> str:
    if not isinstance(texto, str):
        return texto
    return unidecode(texto).lower().strip()

df_municipios["EntidadFederativa"] = df_municipios["EntidadFederativa"].apply(limpiar_texto)
df_municipios["Municipio"]         = df_municipios["Municipio"].apply(limpiar_texto)

# ✅ Verificar que funcionó
print(df_municipios["EntidadFederativa"].unique())
print(df_municipios["Municipio"].head(10))
```

```python
entidades_unicas = df_municipios["EntidadFederativa"].unique()
print(f"Entidades únicas: {len(entidades_unicas)}")
print(entidades_unicas)
```

```python
mapa_entidades = {
    entidad: str(i + 1).zfill(2)
    for i, entidad in enumerate(entidades_unicas)
}
```

```python
for k, v in mapa_entidades.items():
    print(f"  {v} → {k}")

df_municipios["ClaveEntidad"] = (
    df_municipios["EntidadFederativa"]
    .map(mapa_entidades)
)
print(df_municipios[["ClaveEntidad", "EntidadFederativa"]].drop_duplicates().head(10))
print(f"\nShape: {df_municipios.shape}")
df_municipios.head(10)
```

```python
df_municipios["ClaveMunicipio"] = (
    df_municipios["ClaveMunicipio"]
    .astype(str)
    .str.zfill(3)   
)

# Verificar
print(df_municipios[["ClaveEntidad", "ClaveMunicipio"]].head(10))
print(df_municipios[df_municipios["ClaveEntidad"] == "26"].head(5))
```

```python
catalogo = {}

for _, row in df_municipios.iterrows():
    entidad_id = row["ClaveEntidad"]
    entidad_name = row["EntidadFederativa"]
    mun_id = row["ClaveMunicipio"]
    mun_name = row["Municipio"]

    if entidad_name not in catalogo:
        catalogo[entidad_name] = {
            "id"    : entidad_id,
            "municipios": {}
        }

    catalogo[entidad_name]['municipios'][mun_name] = {
        "id"    : mun_id
    }
```

```python
catalogo
```

```python
OUTPUT_JSON = GAS_DATA / Path("catalogo_municipios.json")
with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
    json.dump(catalogo, f, indent=2, ensure_ascii=False)
```
